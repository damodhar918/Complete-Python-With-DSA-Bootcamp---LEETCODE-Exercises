## 1. Factorial Calculation

```python
import math
import time

print("=" * 50)
print("FACTORIAL CALCULATION")
print("=" * 50)

def factorial(n):
    """Calculate factorial of n"""
    if n < 0:
        raise ValueError("Factorial not defined for negative numbers")
    if n == 0 or n == 1:
        return 1
    
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

def factorial_iterative(n):
    """CPU-intensive version - more work per calculation"""
    count = 0
    result = 1
    for i in range(2, n + 1):
        for j in range(1000):  # Simulate more work
            count += 1
        result *= i
    return result, count

# Test basic factorial
print(f"5! = {factorial(5)}")
print(f"10! = {factorial(10)}")
print(f"20! = {factorial(20)}")
```

## 2. Sequential Processing

import time
from math import factorial

print("\n" + "=" * 50)
print("SEQUENTIAL PROCESSING")
print("=" * 50)

numbers = [10000, 10000, 10000, 10000]

def heavy_factorial(n):
    """CPU-intensive factorial simulation"""
    count = 0
    for _ in range(n):
        for i in range(100):
            count += i
    return count

print("Sequential calculation...")
start = time.time()
results = []
for n in numbers:
    result = heavy_factorial(n)
    results.append(result)
    print(f"Processed: {result}")
seq_time = time.time() - start
print(f"Sequential time: {seq_time:.2f}s")
