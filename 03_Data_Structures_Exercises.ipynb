## Coding Exercise 21: Sum of List Elements

Write a function to calculate the sum of all elements in a list.

**Example:**
- Input: [1, 2, 3, 4, 5], Output: 15
- Input: [10, 20, 30], Output: 60
- Input: [-1, 2, -3, 4], Output: 2

```python
# Exercise 21: Sum of List Elements
def sum_of_list(lst):
    """
    Calculate the sum of all elements in a list
    
    Args:
        lst (list): List of numbers
    
    Returns:
        int/float: Sum of all elements
    """
    return sum(lst)

# Test
test_cases = [[1, 2, 3, 4, 5], [10, 20, 30], [-1, 2, -3, 4]]
for test_list in test_cases:
    result = sum_of_list(test_list)
    print(f"Input: {test_list}, Output: {result}")
```

## Coding Exercise 23: Remove Duplicate in a List

Write a function to remove duplicate elements from a list.

**Example:**
- Input: [1, 2, 2, 3, 3, 3], Output: [1, 2, 3]
- Input: [5, 5, 5], Output: [5]
- Input: [1, 2, 3, 4], Output: [1, 2, 3, 4]

```python
# Exercise 23: Remove Duplicate in a List
def remove_duplicates(lst):
    """
    Remove duplicate elements from a list
    
    Args:
        lst (list): List with possible duplicates
    
    Returns:
        list: List without duplicates
    """
    return list(dict.fromkeys(lst))

# Test
test_cases = [[1, 2, 2, 3, 3, 3], [5, 5, 5], [1, 2, 3, 4]]
for test_list in test_cases:
    result = remove_duplicates(test_list)
    print(f"Input: {test_list}, Output: {result}")
```

## Coding Exercise 24: Check if all elements in a list are Unique

Write a function to check if all elements in a list are unique (no duplicates).

**Example:**
- Input: [1, 2, 3, 4], Output: True
- Input: [1, 2, 2, 3], Output: False
- Input: [5], Output: True

```python
# Exercise 24: Check if all elements in a list are Unique
def are_all_unique(lst):
    """
    Check if all elements in a list are unique
    
    Args:
        lst (list): List to check
    
    Returns:
        bool: True if all unique, False otherwise
    """
    return len(lst) == len(set(lst))

# Test
test_cases = [[1, 2, 3, 4], [1, 2, 2, 3], [5]]
for test_list in test_cases:
    result = are_all_unique(test_list)
    print(f"Input: {test_list}, Output: {result}")
```

## Coding Exercise 25: Program to Reverse a List

Write a function to reverse a list.

- Input: ['a', 'b', 'c'], Output: ['c', 'b', 'a']

```python
# Exercise 25: Program to Reverse a List
def reverse_list(lst):
    """
    Reverse a list
    
    Args:
        lst (list): List to reverse
    
    Returns:
        list: Reversed list
    """
    return lst[::-1]

# Test
test_cases = [[1, 2, 3, 4, 5], ['a', 'b', 'c'], [10]]
for test_list in test_cases:
    result = reverse_list(test_list)
    print(f"Input: {test_list}, Output: {result}")
```

## Coding Exercise 26: Count Number of Odd and Even Elements in a List

Write a function to count the number of odd and even elements in a list.

**Example:**
- Input: [1, 2, 3, 4, 5], Output: {'odd': 3, 'even': 2}
- Input: [2, 4, 6, 8], Output: {'odd': 0, 'even': 4}
- Input: [1, 3, 5, 7], Output: {'odd': 4, 'even': 0}

```python
# Exercise 26: Count Number of Odd and Even Elements in a List
def count_odd_even(lst):
    """
    Count odd and even elements in a list
    
    Args:
        lst (list): List of numbers
    
    Returns:
        dict: Dictionary with 'odd' and 'even' counts
    """
    odd_count = sum(1 for num in lst if num % 2 != 0)
    even_count = len(lst) - odd_count
    return {'odd': odd_count, 'even': even_count}

# Test
test_cases = [[1, 2, 3, 4, 5], [2, 4, 6, 8], [1, 3, 5, 7]]
for test_list in test_cases:
    result = count_odd_even(test_list)
    print(f"Input: {test_list}, Output: {result}")
```

## Coding Exercise 27: Maximum difference between two consecutive elements in a list

Write a function to find the maximum difference between two consecutive elements in a list.

**Example:**
- Input: [1, 5, 10, 3, 8], Output: 7 (difference between 10 and 3)
- Input: [2, 4, 6, 8], Output: 2
- Input: [10, 5, 3], Output: 5 (difference between 10 and 5)

```python
# Exercise 27: Maximum difference between two consecutive elements in a list
def max_consecutive_diff(lst):
    """
    Find maximum difference between consecutive elements
    
    Args:
        lst (list): List of numbers
    
    Returns:
        int/float: Maximum difference
    """
    if len(lst) < 2:
        return 0
    return max(abs(lst[i] - lst[i+1]) for i in range(len(lst)-1))

# Test
test_cases = [[1, 5, 10, 3, 8], [2, 4, 6, 8], [10, 5, 3]]
for test_list in test_cases:
    result = max_consecutive_diff(test_list)
    print(f"Input: {test_list}, Output: {result}")
```

## Coding Exercise 28: Merge two Sorted Lists

Write a function to merge two sorted lists into a single sorted list.

**Example:**
- Input: [1, 3, 5], [2, 4, 6], Output: [1, 2, 3, 4, 5, 6]
- Input: [1, 2], [3, 4], Output: [1, 2, 3, 4]
- Input: [5], [1, 2], Output: [1, 2, 5]

```python
# Exercise 28: Merge two Sorted Lists
def merge_sorted_lists(lst1, lst2):
    """
    Merge two sorted lists
    
    Args:
        lst1 (list): First sorted list
        lst2 (list): Second sorted list
    
    Returns:
        list: Merged and sorted list
    """
    return sorted(lst1 + lst2)

# Test
test_cases = [([1, 3, 5], [2, 4, 6]), ([1, 2], [3, 4]), ([5], [1, 2])]
for lst1, lst2 in test_cases:
    result = merge_sorted_lists(lst1, lst2)
    print(f"Input: {lst1}, {lst2}, Output: {result}")
```

## Coding Exercise 29: Rotate a List

Write a function to rotate a list by k positions to the right.

**Example:**
- Input: [1, 2, 3, 4, 5], k=2, Output: [4, 5, 1, 2, 3]
- Input: ['a', 'b', 'c'], k=1, Output: ['c', 'a', 'b']
- Input: [10, 20, 30], k=3, Output: [10, 20, 30]

```python
# Exercise 29: Rotate a List
def rotate_list(lst, k):
    """
    Rotate a list by k positions to the right
    
    Args:
        lst (list): List to rotate
        k (int): Number of positions to rotate
    
    Returns:
        list: Rotated list
    """
    k = k % len(lst)  # Handle k greater than list length
    return lst[-k:] + lst[:-k] if k else lst

# Test
test_cases = [([1, 2, 3, 4, 5], 2), (['a', 'b', 'c'], 1), ([10, 20, 30], 3)]
for test_list, k in test_cases:
    result = rotate_list(test_list, k)
    print(f"Input: {test_list}, k={k}, Output: {result}")
```

## Coding Exercise 30: Merge 2 Lists into Dictionary

Write a function to merge two lists into a dictionary (one as keys, one as values).

**Example:**
- Input: [1, 2, 3], ['a', 'b', 'c'], Output: {1: 'a', 2: 'b', 3: 'c'}
- Input: ['name', 'age'], ['John', 25], Output: {'name': 'John', 'age': 25}
- Input: [1, 2], [10, 20], Output: {1: 10, 2: 20}

```python
# Exercise 30: Merge 2 Lists into Dictionary
def merge_lists_to_dict(keys, values):
    """
    Merge two lists into a dictionary
    
    Args:
        keys (list): List of keys
        values (list): List of values
    
    Returns:
        dict: Merged dictionary
    """
    
    return dict(zip(keys, values)) if len(keys) == len(values) else False

# Test
test_cases = [([1, 2, 3], ['a', 'b', 'c']), (['name', 'age'], ['John', 25]), ([1, 2], [10, 20]), ([1],[2,3])]
for keys, values in test_cases:
    result = merge_lists_to_dict(keys, values)
    print(f"Input: keys={keys}, values={values}, Output: {result}")
```

## Coding Exercise 31: Merge Multiple Dictionaries

Write a function to merge multiple dictionaries into a single dictionary.

**Example:**
- Input: [{'a': 1}, {'b': 2}, {'c': 3}], Output: {'a': 1, 'b': 2, 'c': 3}
- Input: [{'x': 10}, {'y': 20}, {'z': 30}], Output: {'x': 10, 'y': 20, 'z': 30}
- Input: [{'name': 'John'}, {'age': 25}], Output: {'name': 'John', 'age': 25}

```python
# Exercise 31: Merge Multiple Dictionaries
def merge_multiple_dicts(dict_list):
    """
    Merge multiple dictionaries into one
    
    Args:
        dict_list (list): List of dictionaries to merge
    
    Returns:
        dict: Merged dictionary
    """
    merged = {}
    for d in dict_list:
        merged.update(d)
    return merged

# Test
test_cases = [
    [{'a': 1}, {'b': 2}, {'c': 3}],
    [{'x': 10}, {'y': 20}, {'z': 30}],
    [{'name': 'John'}, {'age': 25}]
]
for dict_list in test_cases:
    result = merge_multiple_dicts(dict_list)
    print(f"Input: {dict_list}, Output: {result}")
```

## Coding Exercise 32: Words Frequency in a Sentence

Write a function to count the frequency of each word in a sentence.

**Example:**
- Input: "hello world hello", Output: {'hello': 2, 'world': 1}
- Input: "python python java python", Output: {'python': 3, 'java': 1}
- Input: "a b c a b a", Output: {'a': 3, 'b': 2, 'c': 1}

```python
# Exercise 32: Words Frequency in a Sentence
def word_frequency(sentence):
    """
    Count frequency of each word in a sentence
    
    Args:
        sentence (str): Input sentence
    
    Returns:
        dict: Dictionary with word frequencies
    """
    words = sentence.lower().split()
    freq = {}
    # Get non-existing key with custom default value
    # print(sample_dict.get('z', 0))  # Output: 0
    for word in words:
        freq[word] = freq.get(word, 0) + 1
    return freq

# Test
test_cases = [
    "hello world hello",
    "python python java python",
    "a b c a b a"
]
for sentence in test_cases:
    result = word_frequency(sentence)
    print(f"Input: '{sentence}', Output: {result}")
```

## Coding Exercise 33: Palindromic Tuple

Write a function to check if a tuple is a palindrome (reads same forwards and backwards).

**Example:**
- Input: (1, 2, 3, 2, 1), Output: True
- Input: (1, 2, 3), Output: False
- Input: ('a', 'b', 'a'), Output: True

```python
# Exercise 33: Palindromic Tuple
def is_palindromic_tuple(tup):
    """
    Check if a tuple is palindromic
    
    Args:
        tup (tuple): Tuple to check
    
    Returns:
        bool: True if palindrome, False otherwise
    """
    return tup == tup[::-1]

# Test
test_cases = [(1, 2, 3, 2, 1), (1, 2, 3), ('a', 'b', 'a')]
for test_tuple in test_cases:
    result = is_palindromic_tuple(test_tuple)
    print(f"Input: {test_tuple}, Output: {result}")
```

## Coding Exercise 34: Merge Dictionaries with Common Keys

Write a function to merge two dictionaries, combining values of common keys into a list.

**Example:**
- Input: {'a': 1, 'b': 2}, {'b': 3, 'c': 4}, Output: {'a': 1, 'b': [2, 3], 'c': 4}
- Input: {'x': 10}, {'x': 20}, Output: {'x': [10, 20]}
- Input: {'p': 1, 'q': 2}, {'r': 3}, Output: {'p': 1, 'q': 2, 'r': 3}

```python
# Exercise 34: Merge Dictionaries with Common Keys
def merge_dicts_with_overlapping_keys(dicts):
    """
    Merge multiple dictionaries, summing values of overlapping keys
    
    Args:
        dicts (list): List of dictionaries to merge
    
    Returns:
        dict: Merged dictionary with summed values for overlapping keys
    """
    result = {}
    for d in dicts:
        for key, value in d.items():
            result[key] = result.get(key, 0) + value
    return result

test_cases = [
    [{'a': 1, 'b': 2}, {'b': 3, 'c': 4}, {'c': 5, 'd': 6}],
    [{'x': 10, 'y': 20}, {'y': 30, 'z': 40}, {'z': 50, 'x': 60}]
]
for dict_list in test_cases:
    result = merge_dicts_with_overlapping_keys(dict_list)
    print(f"Input: {dict_list}, Output: {result}")
```

## Coding Exercise 35: Check if List is Subset of another List

Write a function to check if one list is a subset of another list.

**Example:**
- Input: [1, 2], [1, 2, 3, 4], Output: True
- Input: [1, 5], [1, 2, 3], Output: False
- Input: [1, 2, 3], [1, 2, 3], Output: True

```python
# Exercise 35: Check if List is Subset of another List
def is_subset(list1, list2):
    """
    Check if list1 is a subset of list2
    
    Args:
        list1 (list): Potential subset
        list2 (list): Potential superset
    
    Returns:
        bool: True if list1 is subset of list2, False otherwise
    """
    return set(list1).issubset(set(list2))

# Test
test_cases = [
    ([1, 2], [1, 2, 3, 4]),
    ([1, 5], [1, 2, 3]),
    ([1, 2, 3], [1, 2, 3])
]
for list1, list2 in test_cases:
    result = is_subset(list1, list2)
    print(f"Input: {list1}, {list2}, Output: {result}")
```
