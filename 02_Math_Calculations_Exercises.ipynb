## Coding Exercise 16: Celsius to Fahrenheit

Write a function to convert temperature from Celsius to Fahrenheit.

**Formula:** F = (C × 9/5) + 32

**Example:** 
- Input: 0, Output: 32.0
- Input: 100, Output: 212.0
- Input: -40, Output: -40.0

```python
# Exercise 16: Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    """
    Convert Celsius to Fahrenheit
    
    Args:
        celsius (float): Temperature in Celsius
    
    Returns:
        float: Temperature in Fahrenheit
    """
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Test
test_values = [0, 100, -40, 37]
for temp_c in test_values:
    temp_f = celsius_to_fahrenheit(temp_c)
    print(f"Input: {temp_c}°C, Output: {temp_f}°F")
```

```python
import matplotlib.pyplot as plt

# Plot: Celsius to Fahrenheit Conversion (using test values)
test_values = [0, 100, -40, 37]
fahrenheit_values = [celsius_to_fahrenheit(c) for c in test_values]

plt.figure(figsize=(10, 6))
plt.plot(test_values, fahrenheit_values, marker='o', linewidth=2, markersize=8, color='red', label='Temperature')
plt.xlabel('Celsius', fontsize=12)
plt.ylabel('Fahrenheit', fontsize=12)
plt.title('Celsius to Fahrenheit Conversion', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# Add value labels on points
for c, f in zip(test_values, fahrenheit_values):
    plt.annotate(f'{f:.1f}', (c, f), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9)

plt.axhline(y=32, color='blue', linestyle='--', alpha=0.5, label='Freezing Point (32°F)')
plt.axvline(x=0, color='green', linestyle='--', alpha=0.5, label='Water Freezes (0°C)')
plt.legend()
plt.tight_layout()
plt.show()
```

## Coding Exercise 17: Area of a Rectangle

Write a function to calculate the area of a rectangle.

**Formula:** Area = length × width

**Example:**
- Input: length=5, width=3, Output: 15
- Input: length=10, width=4, Output: 40
- Input: length=7.5, width=2.5, Output: 18.75

```python
# Exercise 17: Area of a Rectangle
def rectangle_area(length, width):
    """
    Calculate the area of a rectangle
    
    Args:
        length (float): Length of the rectangle
        width (float): Width of the rectangle
    
    Returns:
        float: Area of the rectangle
    """
    area = length * width
    return area

# Test
test_cases = [(5, 3), (10, 4), (7.5, 2.5)]
for length, width in test_cases:
    area = rectangle_area(length, width)
    print(f"Input: length={length}, width={width}, Output: {area}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot: Rectangle Area (using test cases)
test_cases = [(5, 3), (10, 4), (7.5, 2.5)]
lengths = [case[0] for case in test_cases]
widths = [case[1] for case in test_cases]
areas = [rectangle_area(l, w) for l, w in test_cases]

plt.figure(figsize=(10, 6))
x_pos = np.arange(len(test_cases))
bars = plt.bar(x_pos, areas, color=['skyblue', 'lightcoral', 'lightgreen'], edgecolor='navy', linewidth=1.5)
plt.xlabel('Test Cases', fontsize=12)
plt.ylabel('Area (sq units)', fontsize=12)
plt.title('Area of Rectangle - Test Cases Comparison', fontsize=14, fontweight='bold')
plt.xticks(x_pos, [f'L={l}, W={w}' for l, w in test_cases])
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for i, (bar, area) in enumerate(zip(bars, areas)):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{area}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Coding Exercise 18: Distance covered by a Vehicle

Write a function to calculate the distance covered by a vehicle.

**Formula:** Distance = Speed × Time

**Example:**
- Input: speed=60, time=2, Output: 120
- Input: speed=80, time=3.5, Output: 280
- Input: speed=45, time=1.5, Output: 67.5

```python
# Exercise 18: Distance covered by a Vehicle
def calculate_distance(speed, time):
    """
    Calculate the distance covered by a vehicle
    
    Args:
        speed (float): Speed of the vehicle in km/h
        time (float): Time in hours
    
    Returns:
        float: Distance covered in km
    """
    distance = speed * time
    return distance

# Test
test_cases = [(60, 2), (80, 3.5), (45, 1.5)]
for speed, time in test_cases:
    distance = calculate_distance(speed, time)
    print(f"Input: speed={speed} km/h, time={time} h, Output: {distance} km")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot: Distance covered by Vehicle (using test cases)
test_cases = [(60, 2), (80, 3.5), (45, 1.5)]
speeds = [case[0] for case in test_cases]
times = [case[1] for case in test_cases]
distances = [calculate_distance(s, t) for s, t in test_cases]

plt.figure(figsize=(10, 6))
x_pos = np.arange(len(test_cases))
bars = plt.bar(x_pos, distances, color=['#3498db', '#e74c3c', '#2ecc71'], edgecolor='black', linewidth=1.5)
plt.xlabel('Test Cases', fontsize=12)
plt.ylabel('Distance (km)', fontsize=12)
plt.title('Distance Covered by Vehicle - Test Cases', fontsize=14, fontweight='bold')
plt.xticks(x_pos, [f'Speed={s}km/h\nTime={t}h' for s, t in test_cases])
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar, distance in zip(bars, distances):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{distance} km', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Coding Exercise 19: Number of Rounds of Lift

Write a function to calculate the number of rounds a lift needs to make to transport all people.

**Formula:** Rounds = ⌈Total People / Lift Capacity⌉ (ceiling division)

**Example:**
- Input: total_people=25, capacity=5, Output: 5
- Input: total_people=100, capacity=15, Output: 7
- Input: total_people=50, capacity=10, Output: 5

```python
# Exercise 19: Number of Rounds of Lift
import math

def lift_rounds(total_people, capacity):
    """
    Calculate the number of rounds needed for a lift to transport all people
    
    Args:
        total_people (int): Total number of people to transport
        capacity (int): Capacity of the lift per round
    
    Returns:
        int: Number of rounds needed
    """
    rounds = math.ceil(total_people / capacity)
    return rounds

# Test
test_cases = [(25, 5), (100, 15), (50, 10)]
for total_people, capacity in test_cases:
    rounds = lift_rounds(total_people, capacity)
    print(f"Input: total_people={total_people}, capacity={capacity}, Output: {rounds} rounds")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot: Lift Rounds (using test cases)
test_cases = [(25, 5), (100, 15), (50, 10)]
total_people = [case[0] for case in test_cases]
capacities = [case[1] for case in test_cases]
rounds = [lift_rounds(p, c) for p, c in test_cases]

plt.figure(figsize=(10, 6))
x_pos = np.arange(len(test_cases))
bars = plt.bar(x_pos, rounds, color=['#9b59b6', '#f39c12', '#1abc9c'], edgecolor='black', linewidth=1.5)
plt.xlabel('Test Cases', fontsize=12)
plt.ylabel('Number of Rounds', fontsize=12)
plt.title('Lift Rounds Required - Test Cases', fontsize=14, fontweight='bold')
plt.xticks(x_pos, [f'People={p}\nCapacity={c}' for p, c in test_cases])
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar, round_count in zip(bars, rounds):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{round_count}', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Coding Exercise 20: Line Equation

Write a function to find the y-coordinate on a line given x and the line equation parameters.

**Formula:** y = mx + c (where m is slope and c is y-intercept)

**Example:**
- Input: x=2, m=3, c=5, Output: 11
- Input: x=0, m=2, c=-4, Output: -4
- Input: x=-1, m=1, c=2, Output: 1

```python
# Exercise 20: Line Equation
def line_equation(x, m, c):
    """
    Calculate y-coordinate on a line using the equation y = mx + c
    
    Args:
        x (float): x-coordinate
        m (float): Slope of the line
        c (float): y-intercept
    
    Returns:
        float: y-coordinate
    """
    y = m * x + c
    return y

# Test
test_cases = [(2, 3, 5), (0, 2, -4), (-1, 1, 2)]
for x, m, c in test_cases:
    y = line_equation(x, m, c)
    print(f"Input: x={x}, m={m}, c={c}, Output: y={y}")

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot: Line Equation (using test cases)
test_cases = [(2, 3, 5), (0, 2, -4), (-1, 1, 2)]
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

colors = ['blue', 'red', 'green']

for idx, (ax, (x_test, m, c), color) in enumerate(zip(axes, test_cases, colors)):
    x_range = np.linspace(-5, 5, 100)
    y_range = [line_equation(x, m, c) for x in x_range]
    
    # Plot the line
    ax.plot(x_range, y_range, linewidth=2.5, color=color, label=f'y = {m}x + {c}')
    
    # Plot the test point
    y_test = line_equation(x_test, m, c)
    ax.plot(x_test, y_test, 'o', markersize=10, color='darkred', label=f'Test: ({x_test}, {y_test})', zorder=5)
    
    # Add grid and axes
    ax.axhline(y=0, color='black', linewidth=0.8, linestyle='-', alpha=0.3)
    ax.axvline(x=0, color='black', linewidth=0.8, linestyle='-', alpha=0.3)
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('x', fontsize=11)
    ax.set_ylabel('y', fontsize=11)
    ax.set_title(f'y = {m}x + {c}', fontsize=12, fontweight='bold')
    ax.legend(fontsize=10)
    ax.set_xlim(-5, 5)
    ax.set_ylim(-10, 15)

plt.suptitle('Line Equations: y = mx + c (with Test Points)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
