## Coding Exercise 43: Reverse a string

Write a function to reverse a string.

**Example:**
- Input: "hello", Output: "olleh"
- Input: "python", Output: "nohtyp"
- Input: "a", Output: "a"

```python
# Exercise 43: Reverse a string
def reverse_string(s):
    """
    Reverse a string
    
    Args:
        s (str): String to reverse
    
    Returns:
        str: Reversed string
    """
    return s[::-1]

# Test
test_cases = ["hello", "python", "a", "DSA", "12345"]
for test_str in test_cases:
    result = reverse_string(test_str)
    print(f"Input: '{test_str}', Output: '{result}'")
```

## Coding Exercise 44: Count Vowels in a string

Write a function to count the number of vowels in a string.

**Example:**
- Input: "hello", Output: 2 (e, o)
- Input: "programming", Output: 3 (o, a, i)
- Input: "aeiou", Output: 5

```python
# Exercise 44: Count Vowels in a string
def count_vowels(s):
    """
    Count the number of vowels in a string
    
    Args:
        s (str): Input string
    
    Returns:
        int: Number of vowels
    """
    vowels = 'aeiouAEIOU'
    return sum(1 for char in s if char in vowels)

# Test
test_cases = ["hello", "programming", "aeiou", "PYTHON", "xyz"]
for test_str in test_cases:
    result = count_vowels(test_str)
    print(f"Input: '{test_str}', Output: {result}")
```

## Coding Exercise 46: Check Palindrome

Write a function to check if a string is a palindrome (reads same forwards and backwards).

**Example:**
- Input: "racecar", Output: True
- Input: "hello", Output: False
- Input: "madam", Output: True

```python
# Exercise 46: Check Palindrome
def is_palindrome(s):
    """
    Check if a string is a palindrome
    
    Args:
        s (str): String to check
    
    Returns:
        bool: True if palindrome, False otherwise
    """
    s = s.lower().replace(" ", "")  # Remove spaces and convert to lowercase
    return s == s[::-1]

# Test
test_cases = ["racecar", "hello", "madam", "A man a plan a canal Panama", "noon"]
for test_str in test_cases:
    result = is_palindrome(test_str)
    print(f"Input: '{test_str}', Output: {result}")
```

## Coding Exercise 48: Remove Duplicates in a string

Write a function to remove duplicate characters from a string.

**Example:**
- Input: "hello", Output: "helo"
- Input: "programming", Output: "progamin"
- Input: "aaa", Output: "a"

```python
# Exercise 48: Remove Duplicates in a string
def remove_duplicates(s):
    """
    Remove duplicate characters from a string
    
    Args:
        s (str): Input string
    
    Returns:
        str: String without duplicates
    """
    return ''.join(dict.fromkeys(s))

# Test
test_cases = ["hello", "programming", "aaa", "abccba", "xyz"]
for test_str in test_cases:
    result = remove_duplicates(test_str)
    print(f"Input: '{test_str}', Output: '{result}'")
```

## Coding Exercise 49: Count consonants in a string

Write a function to count the number of consonants in a string.

**Example:**
- Input: "hello", Output: 3 (h, l, l)
- Input: "programming", Output: 8 (p, r, g, r, m, m, n, g)
- Input: "aeiou", Output: 0

```python
# Exercise 49: Count consonants in a string
def count_consonants(s):
    """
    Count the number of consonants in a string
    
    Args:
        s (str): Input string
    
    Returns:
        int: Number of consonants
    """
    vowels = 'aeiouAEIOU'
    return sum(1 for char in s if char.isalpha() and char not in vowels)

# Test
test_cases = ["hello", "programming", "aeiou", "PYTHON", "xyz"]
for test_str in test_cases:
    result = count_consonants(test_str)
    print(f"Input: '{test_str}', Output: {result}")
```

## Coding Exercise 50: Check for anagrams

Write a function to check if two strings are anagrams of each other.

**Logic:** Two strings are anagrams if they contain the same characters with the same frequencies (ignoring spaces and case).

**Example:**
- Input: "listen", "silent", Output: True
- Input: "hello", "world", Output: False
- Input: "abc", "bca", Output: True

```python
# Exercise 50: Check for anagrams
def are_anagrams(s1, s2):
    """
    Check if two strings are anagrams
    
    Args:
        s1 (str): First string
        s2 (str): Second string
    
    Returns:
        bool: True if anagrams, False otherwise
    """
    s1 = s1.lower().replace(" ", "")
    s2 = s2.lower().replace(" ", "")
    return sorted(s1) == sorted(s2)

# Test
test_cases = [("listen", "silent"), ("hello", "world"), ("abc", "bca"), ("The Eyes", "They See")]
for s1, s2 in test_cases:
    result = are_anagrams(s1, s2)
    print(f"Input: '{s1}', '{s2}', Output: {result}")
```

## Coding Exercise 51: Check Subsequence

Write a function to check if one string is a subsequence of another string.

**Logic:** A subsequence is formed by deleting some characters without changing the order of remaining characters.

**Example:**
- Input: "ace", "abcde", Output: True (a-c-e)
- Input: "aec", "abcde", Output: False
- Input: "abc", "abc", Output: True

```python
# Exercise 51: Check Subsequence
def is_subsequence(s, t):
    """
    Check if t is a subsequence of s
    
    Args:
        s (str): String to check in
        t (str): Potential subsequence
    
    Returns:
        bool: True if t is subsequence of s, False otherwise
    """
    it = iter(s)
    return all(c in it for c in t)

# Test
test_cases = [("abcde", "ace"), ("abcde", "aec"), ("abc", "abc"), ("abcd", "bd")]
for s, t in test_cases:
    result = is_subsequence(s, t)
    print(f"Input: s = '{s}', t = '{t}', Output: {result}")
```

```python
def is_subsequence(s, t):
    """
    Function to check if t is a subsequence of s without using built-in functions.
    
    Parameters:
    s (str): The original string.
    t (str): The target subsequence string.
    
    Returns:
    bool: True if t is a subsequence of s, False otherwise.
    """
    i, j = 0, 0  # Initialize two pointers for s and t
    
    # Loop through the original string s
    while i < len(s) and j < len(t):
        # If the characters match, move the pointer for t
        if s[i] == t[j]:
            j += 1
        # Always move the pointer for s
        i += 1
    
    # If we've matched all characters of t, return True
    return j == len(t)


# Test
test_cases = [("abcde", "ace"), ("abcde", "aec"), ("abc", "abc"), ("abcd", "bd")]
for s, t in test_cases:
    result = is_subsequence(s, t)
    print(f"Input: s = '{s}', t = '{t}', Output: {result}")
```

## Coding Exercise 52: Check for Substring

Write a function to check if a string is a substring of another string.

**Example:**
- Input: "lo", "hello", Output: True
- Input: "xy", "hello", Output: False
- Input: "hello", "hello world", Output: True

```python
# Exercise 52: Check for Substring
def is_substring(s1, s2):
    """
    Check if s1 is a substring of s2
    
    Args:
        s1 (str): Substring to find
        s2 (str): String to search in
    
    Returns:
        bool: True if s1 is substring of s2, False otherwise
    """
    return s1 in s2

# Test
test_cases = [("lo", "hello"), ("xy", "hello"), ("hello", "hello world"), ("world", "hello world")]
for s1, s2 in test_cases:
    result = is_substring(s1, s2)
    print(f"Input: '{s1}', '{s2}', Output: {result}")
```

## Coding Exercise 53: Length of the Longest Word

Write a function to find the length of the longest word in a string.

**Example:**
- Input: "hello world", Output: 5 ("hello" and "world")
- Input: "python programming language", Output: 11 ("programming")
- Input: "a ab abc", Output: 3 ("abc")

```python
# Exercise 53: Length of the Longest Word
def longest_word_length(s):
    """
    Find the length of the longest word in a string
    
    Args:
        s (str): Input string
    
    Returns:
        int: Length of the longest word
    """
    words = s.split()
    return max(len(word) for word in words) if words else 0

# Test
test_cases = ["hello world", "python programming language", "a ab abc", "supercalifragilisticexpialidocious", ""]
for test_str in test_cases:
    result = longest_word_length(test_str)
    print(f"Input: '{test_str}', Output: {result}")
```
